```python
#Web scraping medication and their reviews information from the "Ask a patient website"
```

```python
# import libraries
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd# specify the url


#collecting proxies
from http_request_randomizer.requests.proxy.requestProxy import RequestProxy
req_proxy = RequestProxy() 
proxies = req_proxy.get_proxy_list()

#debugging
print(proxies[0].get_address())
print(proxies[0].country)
```

```python
#Fetching the medications starting with letter 'A'

PROXY = proxies[117].get_address()
webdriver.DesiredCapabilities.FIREFOX['proxy']={
    "httpProxy":PROXY,
    "ftpProxy":PROXY,
    "sslProxy":PROXY,
    
    "proxyType":"MANUAL",
    
}
driver = webdriver.Firefox()

urlpage="https://www.askapatient.com/drugalpha.asp?letter=A"
# get web page
driver.get(urlpage)
# execute script to scroll down the page
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
# sleep for 30s
time.sleep(30)
# driver.quit()
```

```python
#Fetching all tr elements
souphandler = BeautifulSoup(driver.page_source, 'lxml')
results=souphandler.find_all('tr')
print(len(results))
```

```python
#Fetching all medicine names

medicines=[]
for tr in results:
    td = tr.find_all('td')
    row = [i.text.replace(u'\xa0', u' ') for i in td]
    if(len(row)==0):
        continue
    medicines.append(row[0])

for medicine in medicines:
    print(medicine)
```
